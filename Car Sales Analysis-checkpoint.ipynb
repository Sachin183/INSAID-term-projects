# 1.3 Important information on Dataset.

The dataframe has three vital aspects: 1. The features of the vehicle in terms of various characteristics such as engine-type, body-style and curb-weight etc., 2. Respective assigned insurance risk rating is in the form of symboling, 3. Normalized losses in use in comparision with other vehicles. 

The symboing corresponds with degree with reference to vehicle being more risky than its price indicates. Vehicles are assigned with risk factor symbol in association with its price. Thus, if the vehicle is risky or less risky, this symbol is could be adjusted accordingly on the scale. Therefore this process is known as "symboling". A value of +3 denotes vehicle is risky and -3 indicates that vehicle is is probably safe. 

Third aspect is a relative average loss payment per insured vehicle year. Such values are normalized for vehicles within a particular size classification (five cylinder big, sports/speciality, luxury vehicle etc...), this represents an average loss per unit of vehicle per year.

# 4.2. Which values are high in the respective dimensions of the data frame inorder to gain insights into it?

```python
def clean_price(price):                                                   # It Creates a new column "clean_price" that replaces out-of-range prices with "NaN"
    if price>=5118.0 and price<=45400.0:                                  # The highest price on record is obseved to be 45400$ (mercedes-benz)
        return price
    else:
        return np.nan
csnew['clean_price'] = csnew['price'].apply(clean_price)

print("'price'")                                                          # Check out the new column and make sure it looks right
print("Minimum value: ", csnew["clean_price"].min())
print("Maximum value: ", csnew["clean_price"].max())
print("How many values are NaN?: ", pd.isnull(csnew['price']).sum())
print("Frequency table for clean_price:")
```
